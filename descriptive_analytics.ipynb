Import Libraries

```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import spearmanr
```

Show plots and other descriptive statistics

```python
def descriptive_statistics(dataset, dataframe):
    sns.histplot(data=dataframe, x='avg_product_rating', hue='helpfulness', element='step')\
        .set(title=f'Histogram of {dataset}: Avg. Product Rating')
    plt.show()
    sns.histplot(data=dataframe, x='richness', hue='helpfulness', element='step')\
        .set(title=f'Histogram of {dataset}: Review Richness')
    plt.show()
    sns.histplot(data=dataframe, x='review_time', hue='helpfulness', element='step')\
        .set(title=f'Histogram of {dataset}: Review Age')
    plt.show()
    sns.histplot(data=dataframe, x='review_length', hue='helpfulness', element='step')\
        .set(title=f'Histogram of {dataset}: Review Length')
    plt.show()
    sns.histplot(data=dataframe, x='helpfulness_ratio').set(title=f'Histogram of {dataset}: Helpfulness Ratio')
    plt.show()
```

Load Data

```python
train_data = pd.read_pickle('Video_Games_final_train.pkl.gz')
dev_data = pd.read_pickle('Video_Games_final_dev.pkl.gz')
test_data = pd.read_pickle('Video_Games_final_test.pkl.gz')
op_data = pd.read_pickle('Office_Products_final_half.pkl.gz')
tg_data = pd.read_pickle('Toys_&_Games_final_half.pkl.gz')

vg_data = pd.concat([train_data, dev_data, test_data])
data = {'Video Games': vg_data, 'Office Products': op_data, 'Toys & Games': tg_data}
```

Run descriptives

```python
num_vars = ('review_time', 'richness', 'avg_product_rating')
for df in data:
    descriptive_statistics(df, data[df])
    for x in num_vars:
        for y in num_vars:
            corr = spearmanr(data[df][x], data[df][y])
            print(f'Correlation of {x} and {y}: {corr[0]:.3f} ({corr[1]:.3f})')
```
